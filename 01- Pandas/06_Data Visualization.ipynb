# Data Visualization

```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load titanic data online from seaborn library
df = sns.load_dataset('titanic')
```

```python
df.plot(kind='scatter', x='age', y='fare', color = 'black')
```

```python
df.groupby('sex')['age'].mean().plot(kind='bar')
plt.ylabel('Average Age')
plt.xlabel('Gender')
```

```python
#plots using seaborn
sns.barplot(data=df, x = 'survived', y = 'fare', hue='class')
```

```python
sns.boxplot(data=df, x = 'survived', y = 'fare', hue='class')
```

```python
sns.scatterplot(data=df, x = 'fare', y = 'age', hue='class')
```

```python
# import pandas and matplotlib.pyplot libraries
import pandas as pd
import matplotlib.pyplot as plt

# create a dictionary with sample data
data = {'Smart Phones': ['Google Pixel', 'I Phone', 'Samsung'],
    'Sales': [100, 150, 80]}

# create a pandas dataframe from the dictionary
df = pd.DataFrame(data)

# plot the dataframe as a bar chart
df.plot(x='Smart Phones', y='Sales', kind='bar', legend=False)

# set the y-axis label
plt.ylabel('Sales')

# set the title of the plot
plt.title('Sales of different Smart Phones')

# save the plot to a file
plt.savefig('bar_chart.png', dpi=300, bbox_inches='tight')

# display the plot
plt.show()
```
