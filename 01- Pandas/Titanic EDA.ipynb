```python
#import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load the dataset
df = sns.load_dataset("titanic")
```

```python
#inspect the dataset
df.head()
```

```python
#check for missing values
df.isnull().sum()
```

```python
#Visualize the missing values
plt.figure(figsize=(12,6))
sns.heatmap(df.isnull(), cbar=False)
plt.title("Missing values heatmap")
plt.show()
```

```python
#Handle missing values 
df['age'] = df['age'].fillna(df['age'].median())
df['embarked'] = df['embarked'].fillna(df['embarked'].mode()[0])
```

```python
df.isnull().sum()
```

```python
#Analyze the distribution of numerical features
for col in ['age', 'fare']:
    plt.figure(figsize=(8,6))
    sns.histplot(data = df, x = col, kde= True)
    plt.title(f"Distribution of {col}")
    plt.show()
```

```python
#analyze the distribtuion of categorical features
for col in ['pclass', 'sex', 'sibsp', 'parch', 'embarked']:
    plt.figure(figsize=(8,6))
    df[col].value_counts().plot(kind='bar')
    plt.title(f"Distibution of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```

```python
#Analyze the relationship between categorical features
plt.figure(figsize=(12,6))
sns.countplot(x = "pclass", hue= "survived", data = df)
plt.title("Survival rate by Passenger class")
plt.xlabel("Passenger class")
plt.ylabel("count")
plt.show()
```

```python
#Analyze the relationship between numeric and categorical features
plt.figure(figsize=(12,8))
sns.boxplot(x = "pclass", hue="age", data = df)
plt.title("Age Distribution by passenger class")
plt.xlabel("Passenger Class")
plt.ylabel("Count")
plt.show()
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv("used_cars_data.csv")
df
```

```python
x = df[['Kilometers_Driven', 'Price', 'Year']]
y = df['Seats']
```

```python
X_train, X_test,Y_train, Y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, Y_train)
```

```python
y_pred = model.predict(X_test)
mse = mean_squared_error(Y_test, y_pred)
r2 = r2_score(Y_test, y_pred)

print(f'Mean Squared Error: {mse:.2f}')
print(f'R-squared: {r2:.2f}')
```

```python
import random
def fibonacci(n):
    if n<=1:
        return n
    else:
        return (fibonacci(n-1) + fibonacci(n-2))
print("Fibonacci Sequence:")
for i in range (10):
    print(fibonacci(i))    
```

```python
class FileManager:
    def __init__ (self, filename, mode):
        self.filename = filename
        self.mode = mode
        self.file = None

    def __enter__(self):
        self.file = open(self.filename, self.mode)
        return self.file
    
    def __exit__(self, exc_type, exc_value, traceback):
        if self.file:
            self.file.close()
with FileManager("output.txt", "w") as file:
    file.write("This is a test")

print("\nFile created successfully!")                
```

```python
import random

# Define the questions and answers
questions = [
    "What is the capital of India?",
    "Which is the largest ocean in the world?",
    "Who was the first president of the United States?",
    "What is the chemical symbol for gold?",
    "In which year did the World War II end?"
]

answers = [
    "New Delhi",
    "Pacific Ocean",
    "George Washington",
    "Au",
    "1945"
]

# Define the prize amounts
prize_amounts = [1000, 5000, 25000, 100000, 500000, 1000000]

def main():
    # Initialize the number of lifelines
    num_lifelines = 3

    print("Welcome to Kaun Banega Crorepati!")
    print("You have", num_lifelines, "lifelines available.")

    current_prize = 0
    question_index = 0

    while True:
        print(f"\nQuestion for {prize_amounts[question_index]}:")
        print(questions[question_index])

        # Ask the user for their answer
        user_answer = input("Your answer: ").lower()

        # Check if the user's answer is correct
        if user_answer == answers[question_index].lower():
            print("Correct answer!")
            current_prize = prize_amounts[question_index]
            question_index += 1

            # Check if the user has reached the final question
            if question_index == len(questions):
                print(f"Congratulations! You have won {current_prize}.")
                break
        else:
            print("Incorrect answer. Better luck next time!")
            break

        # Ask the user if they want to use a lifeline
        if num_lifelines > 0:
            use_lifeline = input("Do you want to use a lifeline? (yes/no) ").lower()
            if use_lifeline == "yes":
                num_lifelines -= 1
                print("Using a lifeline...")
                print("50/50 lifeline:")

                # Remove two incorrect answers
                remaining_answers = [i for i in range(4) if i != random.randint(0, 3)]
                print(f"The possible answers are: {answers[remaining_answers[0]]}, {answers[remaining_answers[1]]}")

                user_answer = input("Your answer: ").lower()
                if user_answer == answers[question_index].lower():
                    print("Correct answer!")
                    current_prize = prize_amounts[question_index]
                    question_index += 1
                else:
                    print("Incorrect answer. Better luck next time!")
                    break
            else:
                print("Okay, let's continue.")
        else:
            print("You have used all your lifelines.")

if __name__ == "__main__":
    main()
```
