**Rural population Mathematics**

```python
#How much rural population increased from 1998 - 2017
pop_rural_1998 = df['POPULATION 1998 (RURAL)']. sum()
pop_rural_2017 = df['ALL SEXES (RURAL)']. sum()
#total poulation from 1998 - 2017
total_population_rural = pop_rural_1998 + pop_rural_2017
print("Total rural population of pakistan from 1998-2017 is: ", total_population_rural)
#change in population
change_in_population = pop_rural_2017 - pop_rural_1998
print("The increase in rural population of Pakistan from 1998 to 2017 is as follows: ", change_in_population)
#Increase in population by percentage
percentage_increase = (pop_rural_1998 / pop_rural_2017) * 100
print("The population increase in 2017 from 1998 by percentage is: ", percentage_increase, "%")
```

**Urban population mathematics**

```python
#How much change in urban population of Pakistan from 1998 - 2017
pop_urban_1998 = df['POPULATION 1998 (URBAN)'].sum()
pop_urban_2017 = df['ALL SEXES (URBAN)'].sum()
#Total Population  of Pakistan from 1998 - 2017
total_population_urban = pop_urban_1998 + pop_urban_2017
print("Total urban population of Pakistan from 1998 - 2017 is: ", total_population_urban)
#Change in population  of Pakistan from 1998 - 2017
change_in_population_urban = pop_urban_2017 - pop_urban_1998
print("Change in rural population of Pakistan from 1998 - 2017 is: ", change_in_population_urban)
#Increase in population  of Pakistan by percentage from 1998 - 2017
percentage_increase_urban = (pop_urban_1998 / pop_urban_2017) * 100
print("The population increase in 2017 from 1998 by percentage is: ", percentage_increase_urban, "%")
```

```python
#Complete population of Pakistan
Complete_population = total_population_rural + total_population_urban
print("The actually complete population of Pakistan from 1998 - 2017 is: ", Complete_population)
```

**Machine Learning Model**

**1. Load the Dataset**

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
# Load iris dataset

iris = load_iris()
df1 = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df1['target'] = iris.target
df1.head()
```

**2. Data preparation**

```python
from sklearn.model_selection import train_test_split

#Features and target
x = df1[iris.feature_names]
y = df1['target']

#split the data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

**3. Model Selection & Training**

```python
from sklearn.tree import DecisionTreeClassifier

#initialize the model
model = DecisionTreeClassifier()

#Train the model
model.fit(x_train, y_train)
```

**4. Model Evaluation**

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

#Make predictions
y_pred = model.predict(x_test)

#Evaluate the model
accuracy  = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

print(f"Accuracy: {accuracy}")
print(f"Classification Report:\n{report}")
print(f"Confusion Matrix: \n{conf_matrix}")
```

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import to_categorical
```

**2. Load and display the Dataset**

```python
Digits = load_digits()
x = Digits.data
y = Digits.target

#onehot encode the target
y = to_categorical(y)

#split the data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

#standardize the data
scalar = StandardScaler()
x_train = scalar.fit_transform(x_train)
x_test = scalar.transform(x_test)
```

**3. Build the neural network model**

```python
#Build the model
model = Sequential()
model.add(Dense(64, input_shape = (x_train.shape[1],), activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(10, activation='softmax'))

#Compile the model
model.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])

#Print the model summary
model.summary()
```

**4. Train the model**

```python
history = model.fit(x_train, y_train, epochs=150, batch_size= 10, validation_split=0.2)
```

**5. Evaluate the model**

```python
loss, accuracy = model.evaluate(x_test, y_test)
print(f"Accuracy: {accuracy}")

#Making predictions
y_pred = model.predict(x_test)
y_pred_classes = np.argmax(y_pred, axis=1)
y_true_classes = np.argmax(y_test, axis=1)

#Confusion matrix and classification report
from sklearn.metrics import confusion_matrix, classification_report

conf1_matrix = confusion_matrix(y_true_classes, y_pred_classes)
report2 = classification_report(y_true_classes, y_pred_classes)

print(f"Confsuion Matrix: \n{conf1_matrix}")
print(f"Classification Report: \n{report2}")
```
