# Plotting with seaborn

```python
#Importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Important things for plotting
1. Data
2. Type of plot
3. x axis
4. y axis
5. Grouping(hue/color) based on categorical variables
6. Size of the plot
7. Color palletes of the plot
8. Style of the plot
9. How do you interpret

```python
#Data
df = sns.load_dataset('titanic')
```

```python
df.info()
```

```python
#Scatter Plot
sns.scatterplot(data= df, x = 'age', y = 'fare', hue='sex')
```

```python
#Scatter Plot
plt.figure(figsize=(10,8))
sns.scatterplot(data= df, x = 'age', y = 'fare', hue='sex')
```

```python
#Scatter Plot
plt.figure(figsize=(10,8))
sns.scatterplot(data= df, x = 'age', y = 'fare', hue='sex')
#change color of scatter points
pallete = {'male': 'black', 'female': 'red'}
sns.scatterplot(data = df, x = 'age', y = 'fare', hue = 'sex', palette=pallete)
#labels
plt.title('Scatter Plot of Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
#saving a plot
plt.savefig('Scatter_Plot.png', dpi = 300)
#saving a plot in vector file
plt.savefig('Scatter_Plot.svg')
```

```python
#Style of a plot
sns.set_theme(style= 'darkgrid', font_scale=1.5)

plt.figure(figsize=(10,8))
sns.scatterplot(data= df, x = 'age', y = 'fare', hue='sex')
#change color of scatter points
pallete = {'male': 'black', 'female': 'red'}
sns.scatterplot(data = df, x = 'age', y = 'fare', hue = 'sex', size='age', palette=pallete)
#labels
plt.title('Scatter Plot of Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
```

Interpertation:
1. We have age of passengers at x axis and fare at y axis.
2. We further grouped the passengers by their genders as male(black points) & female(red points).
3. We can see that the fare is higher for male passengers than female passengers.
4. Two male passengers gave the highest fare.
5. The oldest person at titanic is also a male, having age of 80 years as shown in above figure.

# Box Plot

```python
#Style of a plot
sns.set_theme(style= 'darkgrid', font_scale=1)

plt.figure(figsize=(10,8))
#sns.boxplot(data= df, x = 'age', y = 'fare', hue='sex')
#change color of scatter points
pallete = {'First': 'black', 'Second': 'red','Third': 'yellow'}
sns.boxenplot(data = df, x = 'sex', y = 'age', hue = 'class', palette=pallete)
#labels
plt.title('Boxen plot of Age, based on gender of class')
plt.xlabel('Gender')
plt.ylabel('Age')
#Change legend position and title
plt.legend(loc = 'upper right', title = 'Class')
#change x ticks
plt.xticks([0,1], ['Male', 'Female'], rotation = 69)
```

```python
# bar plot using seaborn
#Style of a plot
sns.set_theme(style= 'darkgrid', font_scale=1)

plt.figure(figsize=(10,8))
#change color of scatter points
pallete = {'First': 'black', 'Second': 'red','Third': 'yellow'}
sns.barplot(data = df, x = 'sex', y = 'age', hue = 'class', palette=pallete, capsize=0.1)
#labels
plt.title('Bar plot of Age, based on gender of class')
plt.xlabel('Gender')
plt.ylabel('Age')
#Change legend position and title
plt.legend(loc = 'upper right', title = 'Class')
#change x ticks
plt.xticks([0,1], ['Male', 'Female'], rotation = 69)
```
